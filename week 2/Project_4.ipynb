```python
#Program to determine the annual tax revenue of various staff based on years of experience and their age
years = int(input("Enter the years of experience of the staff: "))
age = int(input("Enter the age of the staff: "))
if years > 25 and age >= 55:
 print("The annual tax revenue of the staff is: 5,600,000")
elif years > 20 and age >= 45:
 print("The annual tax revenue of the staff is: 4,480,000")
elif years > 10 and age >= 35:
 print("The annual tax revenue of the staff is: 1,500,000")
else:
 print("The annual tax revenue of the staff is: 550,000")
```
